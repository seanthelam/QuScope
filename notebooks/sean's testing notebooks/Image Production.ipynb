```python
mno = read("ICSD_CollCode59159-K1_33Mn8O16.cif")
mno = mno * (5,5,10)
```

```python
def replace_mn_with_ir(mno, fraction):
    new_mno = mno.copy()
    mn_indices = [i for i, atom in enumerate(new_mno) if atom.symbol =="Mn"]
    n_replace = int(len(mn_indices) * fraction)
    ir_indices = random.sample(mn_indices, n_replace)
    for i in ir_indices:
        new_mno[i].symbol = 'Ir'
    return new_mno
```

```python
fig, axs = plt.subplots(3, 2, figsize=(12, 12))

abtem.show_atoms(mno_70_ir,legend=True, title="Beam view (Ir 70%)", ax=axs[0, 0])
abtem.show_atoms(mno_70_ir, plane="xz", title="Side view (Ir 70%)", ax=axs[0, 1])

plt.tight_layout()
plt.show()
```

```python
from ase.io import write
write('Images/Ir-KMnO2_62025_noise/mno_70_ir.xyz', mno_70_ir)
```

```python
potential = abtem.Potential(mno_70_ir, gpts=800, slice_thickness=2.87)

probe = abtem.Probe(energy=200e3, semiangle_cutoff=20.6)
probe.grid.match(potential)

fig, (ax1, ax2) = plt.subplots(1,2)
abtem.show_atoms(mno_70_ir, ax=ax1)
abtem.show_atoms(mno_70_ir, legend=True, ax=ax2, plane="yz")
plt.tight_layout()
```

```python
probe.show(power=0.25);
```

```python
potential.project().show();
```

```python
bf_detector = abtem.AnnularDetector(inner=0, outer=30)
adf_detector = abtem.AnnularDetector(inner=90, outer=200)
pixelated_detector = abtem.PixelatedDetector(max_angle=100)

all_detectors=[bf_detector, adf_detector, pixelated_detector]
```

```python
sampling = probe.ctf.nyquist_sampling
print(f"Nyquist sampling = {sampling:.3f} /pixel")

grid_scan = abtem.GridScan(
    start=[8, 8],
    end= [potential.extent[0]-8,potential.extent[1]-8],
    sampling=sampling,

)

abtem.show_atoms(mno_70_ir,scans=grid_scan);
```

```python
bf, adf, pixelated_measurement = probe.scan(potential=potential,
                     scan=grid_scan,
                     detectors=all_detectors,
                    )
```

```python
from scipy.ndimage import gaussian_filter

def apply_gaussian_blur(image, sigma):
    return gaussian_filter(image, sigma=sigma)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(8, 5))

# Apply Gaussian blur to the images
bf_blurred = apply_gaussian_blur(bf.array, sigma=1.2)
adf_blurred = apply_gaussian_blur(adf.array, sigma=1.2)

# Display the images
ax1.imshow(bf_blurred, cmap='inferno')
ax1.set_title("bright field")
ax1.set_xlabel("x [nm]")
ax1.set_ylabel("y [nm]")

ax2.imshow(adf_blurred, cmap='inferno', vmin=0, vmax = 0.001)
ax2.set_title("dark field")
ax2.set_xlabel("x [nm]")
ax2.set_ylabel("y [nm]")

fig.tight_layout()
# plt.colorbar(ax=ax1, mappable=ax1.images[0], orientation='vertical', label='Intensity')
# plt.colorbar(ax=ax2, mappable=ax2.images[0], orientation='vertical', label='Intensity')
plt.show()
```

```python
dose_per_area=1e4
np.random.seed(100)

fig, (ax1, ax2) = plt.subplots(1,2,figsize=(8,5))

# apply poisson noise to the images
bf_noise = np.random.poisson(bf_blurred * dose_per_area) / dose_per_area
adf_noise = np.random.poisson(adf_blurred * dose_per_area) / dose_per_area

# display images
ax1.imshow(bf_noise, cmap='inferno')
ax1.set_title("bright field")
ax1.set_xlabel("x [nm]")
ax1.set_ylabel("y [nm]")

ax2.imshow(adf_noise, cmap='inferno', vmin=0, vmax = 0.001)
ax2.set_title("dark field")
ax2.set_xlabel("x [nm]")
ax2.set_ylabel("y [nm]")

fig.tight_layout()
# plt.colorbar(ax=ax1, mappable=ax1.images[0], orientation='vertical', label='Intensity')
# plt.colorbar(ax=ax2, mappable=ax2.images[0], orientation='vertical', label='Intensity')
plt.show()
```

```python
from PIL import Image
norm_img = (bf_noise - bf_noise.min()) / (bf_noise.max() - bf_noise.min())
uint8_img = (norm_img * 255).astype(np.uint8)
Image.fromarray(uint8_img).save("Images/Ir-KMnO2_62025_noise/resized_bf_image_Ir70pc_1e4dpa.png")

norm_img = (adf_noise - adf_noise.min()) / (adf_noise.max() - adf_noise.min())
uint8_img = (norm_img * 255).astype(np.uint8)
Image.fromarray(uint8_img).save("Images/Ir-KMnO2_62025_noise/resized_adf_image_Ir70pc_1e4dpa.png")
```
